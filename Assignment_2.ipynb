```python
# replace incorrect values

#1. in feature diabetes_mellitus, replace "\tno" to 'no' and "\tyes" to 'yes'
replace_dict = {'tno': 'no', 'tyes': 'yes'}
db['diabetes mellitus'].replace(replace_dict, inplace=True)

#2 in feature coronary_artery_disease, replace "\tno" to 'no'
replace_dict1 = {'tno': 'no'}
db['coronary artery disease'].replace(replace_dict1, inplace=True)
#3 in feature class, replace 'ckd\t' to 'ckd', 'notckd' to 'not ckd'
replace_dict2 = {'ckd\t': 'ckd', 'notckd': 'not ckd'}
db['class'].replace(replace_dict2, inplace=True)
db
#note that ckd means chronic kidney disease
```

```python
num_cols1 = db.select_dtypes(include=['number']).columns
num_cols1
```

```python
# fill null values in numerical columns : use mean of each feature if where value count is less than or equal to 40 and median if value count is greater than 40
threshold = 40

for col in num_cols1:
    
    non_null_count = db[col].count()

    if non_null_count <= threshold:
        
        mean_value = db[col].mean()
        db[col].fillna(mean_value, inplace=True)
    else:
        
        median_value = db[col].median()
        db[col].fillna(median_value, inplace=True)




```

```python
#convert all categorical values to numeric
db = pd.get_dummies(db, columns=cat_cols)
```

```python
#fill null values in categorical columns by using the mode of each column
#do this using functions
def fill_categorical_nulls_with_mode(db):
    for col in db.columns:
        if db[col].dtype == 'object': 
            mode_value = db[col].mode().iloc[0]  
            db[col].fillna(mode_value, inplace=True) 
fill_categorical_nulls_with_mode(db)
print(db)



```
